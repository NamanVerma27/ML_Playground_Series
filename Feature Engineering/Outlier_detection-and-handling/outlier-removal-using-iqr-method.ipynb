# Distribution Plot of cgpa and placement_exam_marks


```python
plt.figure(figsize=(16,5))
plt.subplot(121)
sns.histplot(df['cgpa'], kde='true', alpha=0.3)

plt.subplot(122)
sns.histplot(df['placement_exam_marks'], kde='true', alpha=0.3)

plt.show()
```

### Observations:
- **CGPA** is following a normal distribution.
- **Placement exam marks** is right-skewe
<p>we can not apply zscore and normal distribution method on "placement_exam_marks
<p>we can apply box-plot IQR method to detect outlier on "placement_exam_marks"</p>

## plotting box-plot of "placement_exam_marks"

```python
sns.boxplot(df['placement_exam_marks'])
```

we can see outliers at the upper end of maximum upperlimit value

To remove them we can apply trimming or capping

### calculating IQR

```python
# Finding the IQR
percentile25= df['placement_exam_marks'].quantile(0.25)
percentile75= df['placement_exam_marks'].quantile(0.75)
iqr= percentile75-percentile25
print("25th percentile: ", percentile25 )
print("75th percentile: ", percentile75 )
print("IQR: ", iqr )
```

```python
# Finding upperlimit and lowerlimit
upperlimit= percentile75 + 1.5*iqr
lowerlimit= percentile25 - 1.5*iqr
print("Upper limit: ", upperlimit )
print("Lower limit: ", lowerlimit )
```

Our min value starts from 0 and that's why we have no outliers on the minimum side , upperlimit is 84.5 , any value greater than that is an outlier

## Finding outlier

```python
df[df['placement_exam_marks']>upperlimit]
# datapoints given below all are outliers acc to iqr method
```

# Handling Technique: Trimming

```python
new_df= df[df['placement_exam_marks']<upperlimit]
```

```python
new_df.shape
```

IN new_df , all outliers got removed

```python
# Comparing
plt.figure(figsize=(16, 8))

# Subplot 1: Histogram of placement_exam_marks
plt.subplot(221)  # Adjusted layout to 2x2 grid
sns.histplot(df['placement_exam_marks'], kde=True, alpha=0.3)

# Subplot 2: Boxplot of placement_exam_marks
plt.subplot(222)
sns.boxplot(df['placement_exam_marks'])

# Subplot 3: Histogram of placement_exam_marks in new_df
plt.subplot(223)
sns.histplot(new_df['placement_exam_marks'], kde=True, alpha=0.3)

# Subplot 4: Boxplot of placement_exam_marks in new_df
plt.subplot(224)
sns.boxplot(new_df['placement_exam_marks'])

# Automatically adjust the spacing
plt.tight_layout()

plt.show()
```

# Handling Technique: Capping

```python
new_df_cap=df.copy()

new_df_cap['placement_exam_marks']= np.where(new_df_cap['placement_exam_marks']>upperlimit, upperlimit, 
                    np.where(new_df_cap['placement_exam_marks']<lowerlimit, lowerlimit, new_df_cap['placement_exam_marks'])
                    )
```

```python
new_df_cap.shape
```

The shape of the dataset is still intact

```python
# Comparing
plt.figure(figsize=(16, 8))

# Subplot 1: Histogram of placement_exam_marks
plt.subplot(221)  # Adjusted layout to 2x2 grid
sns.histplot(df['placement_exam_marks'], kde=True, alpha=0.3)

# Subplot 2: Boxplot of placement_exam_marks
plt.subplot(222)
sns.boxplot(df['placement_exam_marks'])

# Subplot 3: Histogram of placement_exam_marks in new_df
plt.subplot(223)
sns.histplot(new_df_cap['placement_exam_marks'], kde=True, alpha=0.3)

# Subplot 4: Boxplot of placement_exam_marks in new_df
plt.subplot(224)
sns.boxplot(new_df_cap['placement_exam_marks'])

# Automatically adjust the spacing
plt.tight_layout()

plt.show()
```
